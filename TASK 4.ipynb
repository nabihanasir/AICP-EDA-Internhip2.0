<b>Q.4: Draw boxplot for Calories vs Category to spot outliers and max calories category<b>

<b>Q.5: Figure out which exact item contains a high quantity for ['Calories', 'Total Fat', 'Carbohydrates',
'Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium
(% Daily Value)', 'Iron (% Daily Value)'. Display summary link this:<b>

```python
cols = data[['Calories', 'Total Fat', 'Carbohydrates','Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 'Iron (% Daily Value)']]
for nutrient in cols.columns:
    max_index = data[nutrient].idxmax()
    item_name = data.loc[max_index, 'Item']
    nutrient_value = data.loc[max_index, nutrient]
    print(f"{nutrient} - {item_name}: {nutrient_value}")
```

<b>Q.7: Draw a horizontal bar graph for items in each category against the calories.
Here is one sample for all items in Beef & Pork, you have to do it for items in each category. Also, write
your observation.
<b>

```python
grouped = data.groupby('Category')

for category, group_data in grouped:
    plt.figure(figsize=(10, 6))  # Adjust figure size if needed
    plt.barh(group_data['Item'], group_data['Calories'])
    plt.xlabel('Calories')
    plt.yticks(range(len(group_data['Item'])), group_data['Item'], fontsize=8)
    plt.ylabel('Item')
    plt.title(f'Calories in Items - {category}')
    plt.tight_layout()
    plt.show()
```
